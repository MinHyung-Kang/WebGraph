```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
root_path = '/dfs/scratch2/dankang/WebGraph/data/2003_webgraph.stats.'
u_root_path = '/dfs/scratch2/dankang/WebGraph/data/2003_webgraph_u.stats.'
```

```python
def read_ccdistr(cctype):
    if cctype == "scc":
        file_path = root_path + "sccdistr"
    else:
        file_path = u_root_path + "wccdistr"
    
    sizes = []
    cnts = []
    with open(file_path, "r") as f:
        for line in f:
            parsed = [int(x) for x in line.split()]
            sizes.append(parsed[0])
            cnts.append(parsed[1])
            
    return sizes, cnts

scc_sizes, scc_cnts = read_ccdistr('scc')
wcc_sizes, wcc_cnts = read_ccdistr('wcc')
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.gca()
ax.scatter(scc_sizes,scc_cnts, facecolors='none', alpha=0.75, edgecolors='red', marker='s', s=48)
ax.scatter(wcc_sizes,wcc_cnts, facecolors='none', alpha=0.75, edgecolors='blue', marker='o', s=48)
ax.set_yscale('symlog')
ax.set_xscale('symlog')
ax.set_ylim(-1, 10e9)
ax.set_xlim(-1,10e7)
ax.legend(["SCC","WCC"], prop={'size': 15})
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 16)
ax.set_xlabel("Size", fontsize= 16)
ax.set_ylabel("Count", fontsize=16)
ax.set_title("SCC and WCC Distribution", fontsize = 24)
```
